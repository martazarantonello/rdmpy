### Aggregate View Demo


This demo's inputs are an incident code and specified date. Specifying the date ensures the analysis is for the correct incident code, as the pair is unique, not the incident code alone. The aims of this analysis are the quantification of incident impacts in terms of total delay minutes, total cancellations and delay serverity for the duration of the incident.

```python
import sys
import os

# Add project root to Python path
parent_dir = os.path.dirname(os.getcwd())
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

# Force reload of rdmpy modules
for module in list(sys.modules):
    if 'rdmpy' in module:
        del sys.modules[module]

from rdmpy.outputs.analysis_tools import aggregate_view_multiday
```

```python
result1 = aggregate_view_multiday(499279, "24-MAY-2024")
print("\nSummary:")
if result1:
    for key, value in result1.items():
        print(f"  {key}: {value}")
print()
```
